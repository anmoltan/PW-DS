All the errors that are associated with arthmetic operations are defined in ArithmeticError class. 
They are:

-> FloatingPointError

-> OverflowError

-> ZeroDivisionError

For ex:

```python
# zerodivision error

try:
    a=1/0
except ArithmeticError as e:
    print(e.__class__)
else:
    print(a)
```

```python
#overflow error
try:
    p=2.0
    for i in range(1,1000):
        p= p**i
        print(p)
except Exception as e:
    print(e.__class__)
```

LookupError is raised when the index or key is not found in the given sequece or list.

It has two subclasses:

->KeyError

->IndexError

```python
# key error
d={1:'anu',2:'janu', 3:'ansh'}

try:
    for i,j in d.items():
        if i==1:
            print(f'move to {d[i+3]}')
except LookupError as e:
    print(e.__class__)
```

```python
#index error

str= 'jbtgnfkd'
try:
    print(str[8])
except LookupError as e:
    print(e.__class__)
```
